```python
# import packages

import scipy.io as sio  # for loading matlab data

import numpy as np      # for numerical libs

from matplotlib.ticker import FuncFormatter # for custom bar plot labels

import matplotlib.pyplot as plt  # for plotting

import matplotlib.pylab as pylab
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (15, 5),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}
pylab.rcParams.update(params)
from statsmodels.tsa import stattools
from statsmodels.graphics import gofplots
from scipy import stats
from statsmodels.tsa import arima_model
from statsmodels.regression import linear_model
```

# load data

```python
data = sio.loadmat('MSFT_20141103.mat')
#data = sio.loadmat('MSFT_20141107.mat')
#data = sio.loadmat('GOOG_20141103.mat')
```

```python
LOB=data['LOB']

t = (np.array((LOB['EventTime'][0][0][:,0]))-3600000*9.5)*1e-3#time in seconds, measured from NASDAQ opening at 9:30am
bid = np.array(LOB['BuyPrice'][0][0]*1e-4)
bidvol = np.array(LOB['BuyVolume'][0][0]*1.0)
ask = np.array(LOB['SellPrice'][0][0]*1e-4)
askvol = np.array(LOB['SellVolume'][0][0]*1.0)
nMO = np.array(LOB['NumberMO'][0][0]*1.0)#total number of MOs submitted in the given 0.1 second interval
vMO = np.array(LOB['VolumeMO'][0][0]*1.0)#total volume of MOs submitted in the given 0.1 second interval
MO=np.array(LOB['MO'][0][0]*1.0)#information about market orders
dt = t[1]-t[0]
print(dt)
print(t[0:10])
print(len(t))
print(len(bid[:,0]))
print(t[-1])
tcur = 10*60*40
```

# Microprice

```python
midprice = 0.5*(bid[:,0]+ask[:,0])
spread = ask[:,0]-bid[:,0]
plt.plot(t[tcur:(tcur+100)],midprice[tcur:(tcur+100)])
plt.plot(t[tcur:(tcur+100)],bid[tcur:(tcur+100),0])
plt.plot(t[tcur:(tcur+100)],ask[tcur:(tcur+100),0])
plt.title('Midprice, best ask, and best bid')
plt.ylabel('price')
plt.xlabel('time (seconds from trading start)')
plt.show()
```

```python
microprice= (bid[:,0]*askvol[:,0]+ask[:,0]*bidvol[:,0])/(bidvol[:,0]+askvol[:,0])
plt.plot(t[tcur:(tcur+100)], microprice[tcur:(tcur+100)])
plt.title('Microprice')
plt.ylabel('price')
plt.xlabel('time (seconds from trading start)')
plt.show()
```

```python
plt.plot(t, microprice-midprice)
plt.title('Microprice - Midprice')
plt.ylabel('diff')
plt.xlabel('time (seconds from trading start)')
plt.show()
```

```python
my_acovf = stattools.acovf(ret_1sec)
plt.bar(range(20),my_acovf[0:20], width=0.5)#autocovariance of microprice returns over 1s
```

# Order flow

```python
OF = np.cumsum(vMO[:,0]-vMO[:,1])#cumulative order flow up to time t, as a function of t
plt.plot(t[(10*60*30):(10*60*(30+11*30))],OF[(10*60*30):(10*60*(30+11*30))])
```

```python
imin = 10*60*30
#imax = 10*60*(30+11*30)
imax = 10*60*(30+10)
t1 = list(t[imin:imax]-t[imin])
OF1 = list(OF[imin:imax]-OF[imin])
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(t1,OF1)
a = intercept
b = slope
print(a)
print(b)
print(p_value)
print(r_value)
```

```python
OF1_centered = [(OF1[i] - t1[i]*b - a) for i in range(len(OF1))]
plt.plot(t1,OF1_centered)
```

```python
my_acf = stattools.acf(OF1_centered)
plt.bar(range(len(my_acf)),my_acf, width=0.5)
#print(my_acf[0:20])
```

```python
my_DF = stattools.adfuller(OF1_centered,1,'c',None)
print(my_DF[1])#pvalue
phi, sigma = linear_model.yule_walker(OF1_centered,order=1)
print(phi)
```

```python
t2 = t1[::10]
OF2 = OF1[::10]
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(t2,OF2)
a = intercept
b = slope
print(a)
print(b)
print(p_value)
print(r_value)
```

```python
OF2_centered = [(OF2[i] - t2[i]*b - a) for i in range(len(OF2))]
```

```python
my_DF = stattools.adfuller(OF2_centered,1,'c',None)
print(my_DF[1])#pvalue
phi, sigma = linear_model.yule_walker(OF2_centered,order=1)
print(phi)
```

```python
h = range(60*3)
OF_pred = [OF2[-1] + b*h[i] + OF2_centered[-1]*(phi**h[i]-1) for i in range(len(h))]
plt.plot(t2,OF2)
plt.plot(t2,[a+b*t2[i] for i in range(len(t2))])
plt.plot([h[i]+t2[-1] for i in range(len(h))],OF_pred)
OF3 = list(OF[(imax-1):(imax+h[-1]*10)]-OF[imin])
plt.plot([h[i]+t2[-1] for i in range(len(OF3[::10]))],OF3[::10])
```

# LO Imbalance

```python
rho = np.array((bidvol[:,0]-askvol[:,0])/(bidvol[:,0]+askvol[:,0]),ndmin=2).T
```

```python
plt.plot(t, rho)
plt.title('Imbalance')
plt.ylabel(r'$\rho$')
plt.xlabel('time (seconds from trading start)')

a = plt.axes([.65, .6, .2, .2])
idx = (t>3600) & (t<=3600+60)
plt.plot( t[idx], rho[idx])
plt.title('a')
plt.ylabel(r'$\rho$')
plt.xlabel(r'$t$')
plt.xticks([3600,3630,3660])

plt.show()
```

# Volume Imbalance at times of Market Orders

```python
rho_MO = (MO[:,3]-MO[:,4])/(MO[:,3]+MO[:,4])
```

```python
plt.hist(rho_MO[MO[:,7]==-1], bins=np.linspace(-1, 1, num=10),alpha=0.5)
plt.hist(rho_MO[MO[:,7]==+1], bins=np.linspace(-1, 1, num=10),alpha=0.5)
plt.title("Hist of imbalance conditional on MO type")
plt.xlabel(r'$\rho$')
plt.ylabel('Freq')
plt.show()
```

```python
plt.hist(rho_MO, bins=np.linspace(-1, 1, num=10),alpha=0.5)
plt.title("Hist of imbalance conditional on MO")
plt.xlabel(r'$\rho$')
plt.ylabel('Freq')
plt.show()
```

```python
# determine the rate of arrival of market orders conditional on order arrival in a specific bin 
def MOActivity(MO,rho,a,b):
    idx_rho = (rho>=a) & (rho<b)
    idx_rho_MO = (rho_MO>=a) & (rho_MO<b)
    idx_rho_MO_b = idx_rho_MO & (MO[:,7]==-1)
    idx_rho_MO_s = idx_rho_MO & (MO[:,7]==+1)
    
    return [np.sum(idx_rho_MO_b), np.sum(idx_rho_MO_s)]/(dt*np.sum(idx_rho))
```

```python
nbands=5
bands = np.array([-1+2*x/(nbands) for x in range(nbands+1)])

MO_activity=np.zeros((nbands,2))
for i in range(nbands):
    MO_activity[i,:] = MOActivity(MO,rho,bands[i],bands[i+1])
MO_activity=MO_activity.T

plt.bar(bands[0:nbands],MO_activity[0,:],width=[(bands[j+1]-bands[j])*0.9 for j in range(nbands)],alpha=0.4)
plt.bar(bands[0:nbands],MO_activity[1,:],width=[(bands[j+1]-bands[j])*0.9 for j in range(nbands)],alpha=0.4)
plt.title(r'Rate of Arrival of MOs conditional on $\rho\in[a,b)$')
plt.xlabel(r'$\rho\in[a,b)$')
plt.ylabel(r'$\lambda$')
plt.show()
```
