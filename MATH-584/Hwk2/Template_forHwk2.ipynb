```python
import pandas as pd
import numpy as np 
import csv
import matplotlib.pyplot as plt
import math
from pandas_datareader import data
import scipy.optimize
from scipy import stats
from scipy.interpolate import InterpolatedUnivariateSpline
```

```python
#use of linear univariate interpolation
Xs = [1,2,3,4]
Y = [x**3 for x in Xs]
spl = InterpolatedUnivariateSpline(Xs, Y, None, [Xs[0], Xs[-1]], 1)
Xd = [i*0.1 for i in range(50)]
plt.plot(Xd,spl(Xd))
plt.scatter(Xs,Y)
plt.show()
```
