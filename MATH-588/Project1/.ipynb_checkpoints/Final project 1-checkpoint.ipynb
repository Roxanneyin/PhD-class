```python
#Problem 2, Part c
#The goal is to compute the V@R given L is distributed by a Normal Distribution or a Studnet-t Distribution (v dfs)
    #with mean mu and variance
#Inputs: (mu,variance,Distribution,alpha)
#Outputs: V@R_alpha(L)

def VAR(mu,variance,Dist_Type,alpha,df):
    if Dist_Type=="Gaussian":
        return mu+variance**(1/2)*scipy.stats.norm.ppf(alpha)
    if Dist_Type=="Student-t":
        return mu+variance**(1/2)*scipy.stats.t.ppf(alpha,df)
    
print(VAR(5,2,"Gaussian",0.95,4))
print(VAR(5,2,"Student-t",0.95,4))
```

### Question 3

a)

```python
# x as the pmf matrix 
#alpha represents the significant level
def ES_pmf(x, alpha):
    x = np.array(x)
    x = x[x[:,0].argsort()]
    sum = 0
    for i in range(len(x[:,1])):
        sum += x[i,1]
        if (alpha <= sum):
            ES = (sum - alpha) * x[i,0]
            for j in range(len(x[:,1]) - i - 1):
                ES += x[i+j+1,0]*x[i+j+1,1]
            return ES
```

b)

```python
#Here a is the given historical data,
#alpha is the quantile, valued between 0 and 1
#npts is number of points involved in computing integral term in ES.
def ES_sim(a,alpha,npts):
    alpha_vec = np.linspace(alpha,1,npts,endpoint=True)
    stepsize = (1-alpha)/(npts-1)
    b = np.zeros(npts);
    for i in range(0,npts):
        b[i]=VaR(a,alpha_vec[i])
    y = stepsize*sum(b)/(1-alpha)
    return y;
```

c)

```python
#Problem 3, Part c
#The goal is to compute the ES of L distributed by those descibed in (2.c.)
#Inputs: (mu,variance,Distribution,alpha)
#Outputs: ES_alpha(L)

def ES(mu,variance,Dist_Type,alpha,df):
    if Dist_Type == "Gaussian":
        return mu+variance**(1/2)*(scipy.stats.norm.pdf(scipy.stats.norm.ppf(alpha)))/(1-alpha)
    if Dist_Type == "Student-t":
        return mu+variance**(1/2)*((scipy.stats.t.pdf(scipy.stats.t.ppf(alpha,df),df))/(1-alpha))*(df+(scipy.stats.t.ppf(alpha,df))**2)/(df-1)
    
print(ES(5,2,"Gaussian",0.95,4))
print(ES(5,2,"Student-t",0.95,4))
```

```python
#5
#c is a number, b and mu are d*1 vector, sigma is d*d matrix, alpha denotes significant level
def Risk_Multi(mu, sigma, c, b, alpha):
    mean_L = - c - np.dot(b.T, mu)
    var_L = np.dot(np.dot(b.T, sigma),b)
    sigma_L = np.sqrt(var_L)
    VaR_L = norm.ppf(1-alpha, mean_L, sigma_L)
    ES_L = mean_L + sigma_L* norm.pdf(norm.ppf(1-alpha))/(1-alpha)
    return VaR_L,ES_L
```

### Question 6

```python
#
#First the quantities involved:
#alpha is the criteria value
#prob is the probability for the sample of 1 out of {0,1}
#nsim is number of simulation needed; nloss is number of loss we want to include in the calculation
#loss_vec is the loss value with the first entry to be loss corresponding to probability 1-p and the
#2nd entry to be the loss corresponding to probability p
#npts is numebr of points that we would use to calculate expected 
def Test_subadd(alpha, prob, nsim, nloss, loss_vec, npts):
    #Simulate all the data
    b = np.zeros((nloss,nsim))
    for i in range(0, nloss):
        b[i,:] = bernoulli.rvs(prob, size=nsim)
    #Now replace all the entries 1 of the above by 
    b[b<1] = loss_vec[0]
    b[b>0] = loss_vec[1]
    #Now investigate about subadditivity of VaR and ES
    #Sum of elements in each column in an array
    loss_total = np.sum(b, axis=0)
    #Create an array to store values of VaR and ES for each sample
    b2_VaR = np.zeros(nloss)
    b2_ES = np.zeros(nloss);
    #VaR for each loss values:
    for i in range(0, nloss):
        b2_VaR[i] = VaR(b[i, :], alpha)
        b2_ES[i] = ES_sim(b[i, :], alpha,npts)
    #The loss for the sum of losses:
    VaR_total = VaR(loss_total, alpha)
    ES_total = ES_sim(loss_total, alpha, npts)
    #Sum of VaR for individual losses:
    total_VaR = np.sum(b2_VaR)
    total_ES = np.sum(b2_ES)
    #Test subadditivity:
    if (VaR_total<=total_VaR and ES_total<=total_ES):
        return("VaR and expected shortfall are both subadditive in this case")
    elif (VaR_total<=total_VaR and ES_total>total_ES):
        return("VaR is subadditive, but expected shortfall is superadditive in this case")
    elif (VaR_total>total_VaR and ES_total<=total_ES ):
        return("VaR is superadditive, but expected shortfall is subadditive in this case")
    else:
        return("Neither VaR nor expected shortfall are subadditive")
```

Comments: with the above function, we can test subadditivity for VaR and ES given significance level, probility for bernoulli distribution, number of simulations etc. Here we use different significance level with the other inputd fixed. We have found out that ES is always subadditive whereas VaR is subadditive only when significance level is low. So we claim that subadditivity properties among two quantities are independent.

```python
#phi needs to be a number between -1 and 1;
#distribution has option normal and t distribution so far; is it possible to have ant distribution of choice?
#variance is a positive number
#nsteps is an integer indicating the number of steps we want for simulation.
#x0 will be the initial value.
###for t distribution, we need variance larger than 1
def AR1(phi, distribution, variance, nsteps, x0):
    #First verify whether the input satisfies out restriction:
    if (phi >= 1 or phi <= -1):
        return("input value of phi should be between -1 and 1")
    if (distribution != "t" and distribution != "normal"):
        return("input distribution can only be t or normal")
    if (variance <= 0):
        return("input variance should be a positive value")
    if (nsteps <= 0):
        return("input nstep should be a positive integer")
    #Generate a list of random variables
    if (distribution == "normal"):
        sd = np.sqrt(variance)
        array_rand = np.random.normal(0, sd, nsteps)
    elif (distribution == "t"):
        df = 2*variance/(variance-1)
        array_rand = np.random.standard_t(df, nsteps)
    #create an array to store the values 
    x = [x0]
    for i in range(nsteps):
        x.append(x[i] + array_rand[i])
    return x 
```

```python
y = AR1(0, "t",2, 50, 0)
```

```python
def autocorrelation_sample(x,h):
    n = len(x)
    x_bar = np.mean(x)
    array_mean_numerator = np.repeat(x_bar, n-h)
    array_mean_denominator = np.repeat(x_bar, n)
    array_1 = np.subtract(x[h:],array_mean_numerator)
    array_2 = np.subtract(x[:(n-h)],array_mean_numerator)
    array_3 = np.subtract(x,array_mean_denominator)
    rho = np.dot(array_1,array_2)/np.dot(array_3,array_3)
    return rho
```
